# <center> $\frac{du}{dt} = x(x-a)(x-2), 0 < a < 2.$ <center/>
<center> It has three steady points, x = 0, x = 2, and x = a. Determine which of them is stable.<center/> 

<center>$t= \frac{1}{2a} \ln{\left|{x} \right|}+
\frac{1}{a^2-2a} \ln{\left|{x -a} \right|+
\frac{a-1}{2} \ln{|x-2|} + C
}$<center/>

```python
def func(a,c,x):
    first = (1/(2*a)) * math.log(abs(x))
    second = (1/(a**2 - 2*a)) * math.log(abs(x - a))
    third = ((a-1)/2) * math.log(abs(x-2))
    anst = first + second + third + c
    return anst
```

<center> $\frac{du}{dt} = x(x-a)(x-2), 0 < a < 2.$ <center/>

```python
x = np.linspace(-2, 5,2000000)
a = 0.5
c = 0
t = [func(a,c,i) for i in x]
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.plot(t,x,'k.', markersize=1)
plt.title(f'a = {a} , C = {c}')
plt.xlabel("t")
plt.ylabel("x")
plt.subplot(1, 2, 2)
plt.plot(t,x,'k.', markersize=1)
plt.plot(t,[a for i in x],'r')
plt.plot(t,[2 for i in x],'r') 
plt.plot(t,[0 for i in x],'r')
plt.title(f'a = {a} , C = {c}')
plt.xlabel("t")
plt.ylabel("x")
```

# <center> $\frac{du}{dt} = x(x-a)(x-2), 0 < a < 2.$ <center/>
<center> It has three steady points, x = 0, x = 2, and x = a. Determine which of them is stable.<center/> 

```python
def ofunc(a,x):
    slope = x*(x-a)*(x-2)
    return slope
```

```python
x = np.linspace(-1000, 1000,200000)
a = 0.5
slope = [ofunc(a,i)*100 for i in x]
```

```python
plt.figure(figsize=(15, 5))
plt.plot(x,slope,'k.', markersize=1)
plt.title(f'a = {a}')
plt.xlabel("x")
plt.ylabel("slope")
```
